```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
```

```python
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
url = "https://www.encyclopedia-titanica.org/cabins.html"
```

```python
data  = requests.get(url).text
print(data)
```

```python
titanic_data = []
```

```python
soup = BeautifulSoup(data, 'html5lib')
```

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    cabin = col[0].text.strip()
    name = col[1].text.strip()
    evidence = col[2].text.strip()
    #comments = col[3].text
    #data = pd.DataFrame(columns={"cabin":cabin, "name":name, "evidence":evidence}) #"comments":comments})
    titanic_data.append({"cabin":cabin, "name":name, "evidence":evidence})
```

```python
df = pd.DataFrame(titanic_data, columns=["cabin", "name", "evidence", "comments"])
```
